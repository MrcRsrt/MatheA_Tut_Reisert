```python
%%manim -v WARNING -ql --media_dir ./manim_media LinearTransformationSceneExample0

from manim import *

# Wiederholung Linearkombinationen
class LinearTransformationSceneExample0(LinearTransformationScene, MovingCameraScene):
    def __init__(self, **kwargs):
        LinearTransformationScene.__init__(
            self,
            show_coordinates=True,
            i_hat_color=PINK,
            j_hat_color=PURPLE,
            **kwargs
        )

    def construct(self):
        self.camera.frame.scale(1.5)
        self.wait()
        vector = Vector([2, 1])
        self.add_vector(vector)
        self.wait()
        matrix = [[2, 0], 
                  [0, 3]]
        self.apply_matrix(matrix)
        self.wait()
        
        # Get their endpoints as coordinates
        i_end = self.i_hat.get_end()
        j_end = self.j_hat.get_end()

        # Add another i_hat, starting at the end of the first one
        second_i = Vector(self.i_hat.get_end() - ORIGIN, color=PINK).shift(i_end)
        self.add_vector(second_i)
        self.wait()

        # Add j_hat, starting at the end of the second i_hat
        third_vec = Vector(self.j_hat.get_end() - ORIGIN, color=PURPLE).shift(2 * i_end)
        self.add_vector(third_vec)
        self.wait()
```

```python
%%manim -v WARNING -ql --media_dir ./manim_media ParallelogramDeterminant

from manim import *
import numpy as np

# Vergroesserung des Flaecheninhalts durch Determinante
class ParallelogramDeterminant(LinearTransformationScene):
    def __init__(self, **kwargs):
        LinearTransformationScene.__init__(
            self,
            show_coordinates=True,
            leave_ghost_vectors=False,
            show_basis_vectors=True,
            **kwargs
        )

    def construct(self):
        # === Matrix hier anpassen ===
        A = np.array([[2, 0],
                      [0, 2]], dtype=float)

        # Basisvektoren
        e1 = np.array([1.0, 0.0, 0.0])
        e2 = np.array([0.0, 1.0, 0.0])

        # Einheitsquadrat
        unit_square = Polygon(ORIGIN, e1, e1 + e2, e2, stroke_width=3)\
            .set_stroke(color=BLUE).set_fill(BLUE, opacity=0.25)

        # Quadrat anzeigen
        self.play(FadeIn(unit_square))
        self.wait(0.5)

        # Transformierte Ecken berechnen
        Ae1 = self._apply_matrix_to_vec(A, e1)
        Ae2 = self._apply_matrix_to_vec(A, e2)

        # Parallelogramm nach Transformation
        transformed_parallelogram = Polygon(ORIGIN, Ae1, Ae1 + Ae2, Ae2, stroke_width=3)\
            .set_stroke(color=GREEN).set_fill(GREEN, opacity=0.25)

        # Transformation (Gitter + Quadrat)
        self.apply_matrix(
            A,
            added_anims=[Transform(unit_square, transformed_parallelogram)],
            run_time=2
        )
        self.wait()

    @staticmethod
    def _apply_matrix_to_vec(A: np.ndarray, v3: np.ndarray) -> np.ndarray:
        x, y = v3[0], v3[1]
        X = A[0,0]*x + A[0,1]*y
        Y = A[1,0]*x + A[1,1]*y
        return np.array([X, Y, 0.0])
```

```python
%%manim -v WARNING -ql --media_dir ./manim_media ParallelogramDeterminant

from manim import *
import numpy as np

#Determinante mit Vorzeichen (positiv)

class ParallelogramDeterminant(LinearTransformationScene):
    def __init__(self, **kwargs):
        LinearTransformationScene.__init__(
            self,
            show_coordinates=True,
            leave_ghost_vectors=False,
            show_basis_vectors=True,
            **kwargs
        )

    def construct(self):
        # === Matrix hier anpassen ===
        A = np.array([[1.5, 1],
                      [1, 2]], dtype=float)

        # Basisvektoren
        e1 = np.array([1.0, 0.0, 0.0])
        e2 = np.array([0.0, 1.0, 0.0])

        # Einheitsquadrat
        unit_square = Polygon(ORIGIN, e1, e1 + e2, e2, stroke_width=3)\
            .set_stroke(color=BLUE).set_fill(BLUE, opacity=0.25)

        # Orientierungspfeil (zeigt e1 -> e2, also positive Orientierung)
        orient_arrow = CurvedArrow(0.6*e1, 0.6*e2, angle=TAU/6, stroke_width=3)\
            .set_color(YELLOW)

        # Startobjekte einblenden
        self.play(FadeIn(unit_square), FadeIn(orient_arrow))
        self.wait(0.5)

        # Transformierte Ecken
        Ae1 = self._apply_matrix_to_vec(A, e1)
        Ae2 = self._apply_matrix_to_vec(A, e2)

        # Parallelogramm nach Transformation
        transformed_parallelogram = Polygon(ORIGIN, Ae1, Ae1 + Ae2, Ae2, stroke_width=3)\
            .set_stroke(color=GREEN).set_fill(GREEN, opacity=0.25)

        # Orientierung nach Transformation (um det(A)-Vorzeichen zu zeigen)
        transformed_orient = CurvedArrow(0.6*Ae1, 0.6*Ae2, angle=TAU/6, stroke_width=3)

        # Farbe je nach Vorzeichen der Determinante
        detA = float(np.linalg.det(A))
        transformed_orient.set_color(RED if detA < 0 else YELLOW)

        # Transformation
        self.apply_matrix(
            A,
            added_anims=[
                Transform(unit_square, transformed_parallelogram),
                Transform(orient_arrow, transformed_orient)
            ],
            run_time=2
        )
        self.wait()

    @staticmethod
    def _apply_matrix_to_vec(A: np.ndarray, v3: np.ndarray) -> np.ndarray:
        """ wendet 2×2-Matrix A auf einen 3D-Vektor v3 (z=0) an """
        x, y = v3[0], v3[1]
        X = A[0,0]*x + A[0,1]*y
        Y = A[1,0]*x + A[1,1]*y
        return np.array([X, Y, 0.0])
```
