```python
# Import necessary modules
from manim import * #Tolle Library für Mathe-Animationen
import numpy as np # Für numerische Operationeny
```

```python


class VectorOperations(Scene):
    def construct(self):
        # ---------- SETTINGS ----------
        # Define two vectors (edit these!)
        v1 = np.array([2, 1, 0])
        v2 = np.array([1, 2, 0])
        scalar = 2  # scalar multiplier for demonstration

        # ---------- SETUP ----------
        plane = NumberPlane(
            x_range=[-1, 5, 1],
            y_range=[-1, 5, 1],
            background_line_style={"stroke_opacity": 0.4}
        )
        self.add(plane)

        # ---------- ORIGINAL VECTORS ----------
        vec1 = Arrow(start=ORIGIN, end=v1, buff=0, color=RED)
        vec2 = Arrow(start=ORIGIN, end=v2, buff=0, color=BLUE)
        label_v1 = MathTex("\\vec{v}_1", color=RED).next_to(vec1.get_end(), DOWN)
        label_v2 = MathTex("\\vec{v}_2", color=BLUE).next_to(vec2.get_end(), LEFT)

        self.play(GrowArrow(vec1), Write(label_v1))
        self.play(GrowArrow(vec2), Write(label_v2))
        self.wait(1)

        # ---------- SCALAR MULTIPLICATION ----------
        self.play(FadeOut(label_v1), FadeOut(label_v2))
        self.wait(0.3)

        scaled_vec1 = Arrow(
            start=ORIGIN, end=scalar * v1, buff=0, color=YELLOW
        )
        label_scaled = MathTex(f"{scalar}\\,\\vec{{v}}_1", color=YELLOW).next_to(
            scaled_vec1.get_end(), RIGHT
        )

        text_scalar = Text("Scalar multiplication", font_size=30).to_edge(UP)
        self.play(Write(text_scalar))
        self.play(Transform(vec1, scaled_vec1), Write(label_scaled))
        self.wait(1.5)

        # ---------- VECTOR ADDITION ----------
        self.play(FadeOut(text_scalar), FadeOut(label_scaled))
        self.wait(0.3)

        # Move v2 to tip of v1 (tip-to-tail)
        moved_v2 = Arrow(start=scalar * v1, end=scalar * v1 + v2, buff=0, color=BLUE)
        result_vec = Arrow(start=ORIGIN, end=scalar * v1 + v2, buff=0, color=GREEN)
        label_result = MathTex("\\vec{v}_1 + \\vec{v}_2", color=GREEN).next_to(
            result_vec.get_end(), RIGHT
        )

        text_addition = Text("Vector addition (tip-to-tail)", font_size=30).to_edge(UP)
        self.play(Write(text_addition))
        self.play(GrowArrow(moved_v2))
        self.play(GrowArrow(result_vec), Write(label_result))
        self.wait(2)
```
